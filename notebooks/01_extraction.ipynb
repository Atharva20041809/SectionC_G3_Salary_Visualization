# Capstone Project: ETL Phase 1 - Extraction
This notebook handles the initial ingestion of the Adult Census Income dataset into our data pipeline.

### Step 1: Environment Setup

```python
import os
import urllib.request
import zipfile
import pandas as pd

# Define paths
DATA_DIR = '../data/raw'
RAW_FILE_PATH = os.path.join(DATA_DIR, 'adult.csv')

# Ensure directory exists
os.makedirs(DATA_DIR, exist_ok=True)
print("Environment dependencies loaded.")
```

### Step 2: Data Extraction
Simulating ingestion from a remote source or a data lake architecture. We assume the raw data comes from a direct URL source or an unzipped format. Since the local infrastructure relies on `adult.csv`, we will verify its existence and structure.

```python
# Capstone Simulation: If we had a remote ZIP source, we would extract it.
# In our existing structure, adult.csv is our definitive raw source.
# Let's perform a validation check to ensure Data Ingestion was successful.

def verify_extraction(file_path):
    if os.path.exists(file_path):
        size_mb = os.path.getsize(file_path) / (1024 * 1024)
        print(f"✅ Extraction Successful: Verified {file_path} (Size: {size_mb:.2f} MB)")
        return True
    else:
        print(f"❌ Extraction Failed: Missing origin file {file_path}")
        return False

extraction_status = verify_extraction(RAW_FILE_PATH)
```

### Step 3: Raw Data Profiling
Quick sanity check to observe the schema and properties of our extracted payload before shipping it to the Cleaning ETL phase.

```python
if extraction_status:
    # Load into DataFrame to confirm ingestion boundaries
    raw_df = pd.read_csv(RAW_FILE_PATH, skipinitialspace=True)
    
    print(f"Dataset Extracted Shape: {raw_df.shape}")
    print("\nColumns Extracted:")
    print(raw_df.columns.tolist())
    
    display(raw_df.head())
else:
    print("Cannot stream schema, file unavailable.")
```
